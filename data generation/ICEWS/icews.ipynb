```python
filtered_df = df[["Source Name", "Event Text", "Target Name", "Event Date", "Intensity", "Publisher", "Country", "City", "Province", "Longitude", "Latitude"]]
```

```python
all_entity= set(\
list(filtered_df['Source Name']) +\
list(filtered_df['Target Name']) + \
list(filtered_df['Publisher']) + \
list(filtered_df['Publisher']) +\
list(filtered_df['Country']) +\
list(filtered_df['City']) +\
list(filtered_df['Province']))


```

```python
all_relation = list(set(list(filtered_df['Event Text']) + ["Intensity", 'Publisher','Country', 'City', 'Province', 'Longitude', 'Latitude']))
```

```python
entity2id = dict()
id2entity = dict()
for idx, entity in enumerate(all_entity):
    entity2id[entity] = idx
    id2entity[idx] = entity
```

```python
relation2id = dict()
id2relation = dict()
for idx, relation in enumerate(all_relation):
    relation2id[relation] = idx
    id2relation[idx] = relation
```

```python
dates = list(filtered_df["Event Date"].drop_duplicates())
```

```python
timestamp2idx = dict()
for idx, t in enumerate(dates):
    timestamp2idx[t] = idx
```

```python
train_idx = int(len(filtered_df) * 0.8)
valid_idx = int(len(filtered_df) * 0.9)
flag = 0
train = open(output_dir + "train.txt" , 'w')
valid = open(output_dir + "valid.txt" , 'w')
test = open(output_dir + "test.txt" , 'w')
with open(output_dir + "data.txt" , 'w') as f:
    for idx, row in tqdm(filtered_df.iterrows()):
        outrow = ""
        outrow += str(entity2id[row["Source Name"]])
        outrow += "\t"
        outrow += str(relation2id[row["Event Text"]])
        outrow += "\t"
        outrow += str(entity2id[row["Target Name"]])
        outrow += "\t"
        
        outrow += str(timestamp2idx[row["Event Date"]])
        outrow += "\t"        
        
        outrow += str(row["Event Date"])
        outrow += "\t"
        
        if row["Event Date"].split("-")[-2] == '11':
            flag = 1 
        elif row["Event Date"].split("-")[-2] == '12':
            flag = 2
        outrow += str(relation2id["Intensity"])
        outrow += "\t"
        outrow += str(row["Intensity"])
        outrow += "\t"
        
        if row["Publisher"] != None:
            outrow += str(relation2id["Publisher"])
            outrow += "\t"
            outrow += str(entity2id[row["Publisher"]])
            outrow += "\t"
            
        if row["Country"] != None:    
            outrow += str(relation2id["Country"])
            outrow += "\t"
            outrow += str(entity2id[row["Country"]])
            outrow += "\t"
            
        if row["City"] != None:
            outrow += str(relation2id["City"])
            outrow += "\t"
            outrow += str(entity2id[row["City"]])
            outrow += "\t"
        
        if row["Province"] != None:
            outrow += str(relation2id["Province"])
            outrow += "\t"
            outrow += str(entity2id[row["Province"]])
            outrow += "\t"
        if row["Longitude"] != None:
            outrow += str(relation2id["Longitude"])
            outrow += "\t"
            outrow += str(row["Longitude"])
            outrow += "\t"   
        if row["Latitude"] != None:
            outrow += str(relation2id["Latitude"])
            outrow += "\t"
            outrow += str(row["Latitude"]) 
            outrow += "\t"
        outrow = outrow[:-1]
        outrow += "\n"
            
#         if idx == train_idx:
#             flag = 1
#         elif idx == valid_idx:
#             flag = 2
            
        if flag == 0 :
            train.write(outrow)
        elif flag == 1:
            valid.write(outrow)
        elif flag == 2:
            test.write(outrow)
        f.write(outrow)
train.close()
valid.close()
test.close()
```

```python
with open(output_dir + "entity2id.txt" , 'w') as f:
    for entity, idx in entity2id.items():
        f.write(str(entity) + "\t" + str(idx) + "\n")
```

```python
with open(output_dir + "relation2id.txt" , 'w') as f:
    for relation, idx in relation2id.items():
        f.write(str(relation) + "\t" + str(idx) + "\n")
```
