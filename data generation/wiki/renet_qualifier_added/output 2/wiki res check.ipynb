```python
with open('data.txt', 'r') as f:
    data = f.read()

data = [i.split("\t") for i in data.split("\n")[:-1]]


import pandas as pd

from collections import Counter

Counter([len(i) for i in data])
```

```python
sorted(Counter([i[1] for i in data]).items(), key = lambda x : -x[1])
```

```python
with open('entity2id.txt', 'r') as f:
    relation2id = f.read()

ent2id = dict()
id2ent = dict()
for k, idx in [i.split("\t") for i in relation2id.split("\n")]:
    ent2id[k] = idx
    id2ent[idx] = k
```

```python
with open('relation2id.txt', 'r') as f:
    relation2id = f.read()

rel2id = dict()
id2rel = dict()
for k, idx in [i.split("\t") for i in relation2id.split("\n")]:
    rel2id[k] = idx
    id2rel[idx] = k
```

```python
import matplotlib.pyplot as plt

### relation
plt.plot(*zip(*sorted(sorted([(k, v) for (k, v) in Counter([id2ent[i[0]] for i in data]).items()][:10]), key = lambda x : -x[1])))
plt.xticks(rotation = 90)
plt.title("subject concentration")
plt.show()
```

```python
import matplotlib.pyplot as plt

### relation
plt.plot(*zip(*sorted(sorted([(k, v) for (k, v) in Counter([id2rel[i[1]] for i in data]).items()][:10]), key = lambda x : -x[1])))
plt.xticks(rotation = 90)
plt.title("relation concentration")
plt.show()
```

```python
count = 0
subject = set()
obj = set()
relation = set()
entity = set()
qk = set()
qv = set()
time = set()
for row in data:
    subject.add(row[0])
    entity.add(row[0])
    relation.add(row[1])
    obj.add(row[2])
    entity.add(row[2])
    time.add(row[3])
    
    for idx, i in enumerate(row[4:]):
        if idx%2 == 0:
            qk.add(i)
        else:
            qv.add(i)
            count +=1
        
print("#sub : ", len(subject))
print("#obj : ", len(obj))
print("#rel : ", len(relation))
print("#ent : ", len(entity))
print("#qk : ", len(qk))
print("#qv : ", len(qv))
print("#time : ", len(time))
    
    
```
